### **SEARCH QUERY**

API search key digunakan untuk mencari keyword "MBKM" yang berbahasa Indonesia dan filter postingan reetweet, maksimal data 100.
Data yang diambil berupa user (nama pengguna) dan text berupa postingan yang diunggah oleh pengguna.

### **MEMBENTUK DATA FRAME**

```python
search_dict = api.search(q="MBKM -filter:retweets", lang="id", count=100)

def tweet(search_dict): 
    tweet_list = {}
    for tweet in search_dict:
        tweet_list[tweet.user.screen_name] = tweet.text
    return tweet_list

mydict = tweet(search_dict)

mydict
```

### **CRAWLING DATA**

Membuat variabel berupa crawling dataset untuk menyimpan data hasil crawling dalam sebuah data frame yang terdiri dari 2 buah kolom berupa username dan komentar

```python
crawling_dataset = pd.DataFrame(list(mydict.items()), columns = ['username', 'komentar'])
crawling_dataset
```
